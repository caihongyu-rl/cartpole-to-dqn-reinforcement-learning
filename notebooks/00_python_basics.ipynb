## Lecture 1: Conditionals

```python
# compare
x=int(input("what's x"))
y=int(input("what's y"))

if x > y:
    print("x is greater than y")

elif x < y:
    print("x is less than y")

else:
    print("x is equal to y")

```

```python
# compare "or"
x=int(input("what's x"))
y=int(input("what's y"))

if x > y or x < y:
    print("x is not equal to y")

else:
    print("x is equal to y")
```

```python
# use while true to create a infinite loop to check the state

while True:
    n = int(input("what's n? "))
    if n > 0:
        break

for _ in range(n):
    print("miao")
```

```python
# create a miao function 
def main():
    number = get_number()
    miao(number)

def get_number():
    while True:
        n = int(input("what's number? "))
        if n > 0:
            break
    return n

def miao(n):
    for _ in range(n):
        print("miao")

main()

```

```python
# mario-square-1

def main():
    print_square(3)

def print_square(size):
    
    # for each row in square
    for i in range(size):

        # for each brick in row
        for j in range(size):

            # print break
            print("#", end = "")

        # after print each brick in row, we need to let the cursor on next line
        print()

main()
```

```python
# mario-square-2

def main():
    print_square(3)

def print_square(size):
    for i in range(size):
        print_row(size)

def print_row(width):
    print("#" * width)

main()
        
```

## Lecture 3: Exceptions

```python
# SyntaxError

print("hello, world)
```

```python
# ValueError

x = int(input("what's x? "))
print(f"x is {x}")
```

```python
# try-except

try:
    x = int(input("what's x? "))
    print(f"x is {x}")
except ValueError:
    print("x is not an integer")
```

```python
# interesting phenomenon-NameError
# int("hello") is wrong, so x don't be evaluation，name 'x' is not defined

try:
    x = int(input("what's x? "))
except ValueError:
    print("x is not an integer")

print(f"x is {x}")
```

```python
# try-except-else
# if the program don't have the error we can execute else,
# if the program have the error we execute except and skip else

try:
    x = int(input("what's x? "))
except ValueError:
    print("x is not an integer")
else:
    print(f"x is {x}")
```

```python
while True:
    try:
        x = int(input("what's x? "))
    except ValueError:
        print("x is not an integer")
    else:
        break

print(f"x is {x}")
```

```python
def main():
    x = get_int()
    print(f"x is {x}")

def get_int():

    while True:
        try:
            x = int(input("what's x? "))
            return x
        except ValueError:
            print("x is not integer")

main()
```
