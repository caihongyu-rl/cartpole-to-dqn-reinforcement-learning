## Lecture 1: Conditionals

```python
# compare
x=int(input("what's x"))
y=int(input("what's y"))

if x > y:
    print("x is greater than y")

elif x < y:
    print("x is less than y")

else:
    print("x is equal to y")

```

```python
# compare "or"
x=int(input("what's x"))
y=int(input("what's y"))

if x > y or x < y:
    print("x is not equal to y")

else:
    print("x is equal to y")
```

```python
# grade "and"
score=int(input("score: "))

if score >=90 and score <= 100: # 90 <= score <= 100
    print("grade: A")

elif score >= 80 and score < 90:
    print("scorez: B")
```

## Lecture 2: loops

```python
# while loop
i=3
while i != 0:
    print("miao")
    i = i - 1
```

```python
i = 0
while i < 3:
    print("miao")
    i += 1
```

```python
# for loop

for i in [0, 1, 2]:
    print("miao")
     
```

```python
# use the range function in for loop

for i in range(3):
    print("miao")
```

```python
# use "_" variable don't care about name in for loop

for _ in range(3):
    print("miao")
```

```python
print("miao\n" * 3, end = "")
```

```python
# use while true to create a infinite loop to check the state

while True:
    n = int(input("what's n? "))
    if n > 0:
        break

for _ in range(n):
    print("miao")
```

```python
# create a miao function 
def main():
    number = get_number()
    miao(number)

def get_number():
    while True:
        n = int(input("what's number? "))
        if n > 0:
            break
    return n

def miao(n):
    for _ in range(n):
        print("miao")

main()

```

```python
# mario-square-1

def main():
    print_square(3)

def print_square(size):
    
    # for each row in square
    for i in range(size):

        # for each brick in row
        for j in range(size):

            # print break
            print("#", end = "")

        # after print each brick in row, we need to let the cursor on next line
        print()

main()
```

```python
# mario-square-2

def main():
    print_square(3)

def print_square(size):
    for i in range(size):
        print_row(size)

def print_row(width):
    print("#" * width)

main()
        
```
